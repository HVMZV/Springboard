```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
wine_tasting = pd.read_csv('C:/Users/hamza/Downloads/Springboard Case Studies/Capstone Project 3/winemag-data-130k-v2.csv')
```

```python
wine_tasting = wine_tasting.set_index('id')
wine_tasting.info()
```

```python
correlation_wine_tasting = wine_tasting.corr()
print(correlation_wine_tasting)
```

This indicates that prices and points have a medium strength, positive relationship.

```python
sns.set(rc={'figure.figsize':(15,10)})

sns.scatterplot(data=wine_tasting, s=40, x=wine_tasting.points, y='price', hue='country')
```

```python
ax = sns.violinplot(x=wine_tasting.index, y="price", data=wine_tasting)
```

In the scatterplots you see what some of the high correlations were clearly picking up on. There's a medium strength positive correlation between price and points. What we can infer from this is that the points score received on wine plays a huge part in determining its price.
